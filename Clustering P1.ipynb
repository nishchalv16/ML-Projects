# Performance Metrics Clustering - Silhoutte Coeff 

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples,silhouette_score

import matplotlib.cm as cm
```

```python
# Generating sample data from make_blobs
##This particular setting has 1 distinct cluster and 3 clusters placed together

X,y = make_blobs(n_samples=500,n_features=2,centers=4,cluster_std=1,center_box=(-10.0,10.0), shuffle=True,random_state=1) # For reproducibility
range_n_clusters = [2,3,4,5,6]
```

```python
X
```

```python
y
```

```python
from sklearn.cluster import KMeans

wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=0) #initialisisng with k-means++ 
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) #gives distance bw cemtroid and inner points
    
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('No of clusters')
plt.ylable('WCSS')
plt.show()
```
